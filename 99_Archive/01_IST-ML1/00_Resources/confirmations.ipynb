# Information gain (calculations)

```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from collections import Counter
import math

# Function to calculate entropy
def entropy(labels):
    label_counts = Counter(labels)
    total_count = len(labels)
    entropy_value = 0.0
    for count in label_counts.values():
        probability = count / total_count
        entropy_value -= probability * math.log2(probability)
    return entropy_value

# Function to calculate information gain
def information_gain(df, feature, target):
    # Calculate entropy of the whole dataset (before split)
    total_entropy = entropy(df[target])
    
    # Calculate the entropy for each split by feature values
    values, counts = np.unique(df[feature], return_counts=True)
    
    # Weighted entropy after the split
    weighted_entropy = 0.0
    for value, count in zip(values, counts):
        subset = df[df[feature] == value]
        subset_entropy = entropy(subset[target])
        weighted_entropy += (count / len(df)) * subset_entropy

    # Information Gain is the difference
    return total_entropy - weighted_entropy

# Step 1: Create the new DataFrame
data = {
    'Weekend': ['Yes', 'No', 'No', 'No', 'Yes'],
    'Weather': ['Sunny', 'Rain', 'Rain', 'Claudy', 'Claudy'],
    'Tired': ['No', 'Yes', 'No', 'No', 'Yes'],
    'What to Do?': ['Go for a walk', 'TV', 'Reading', 'Go for a walk', 'Reading']
}

df = pd.DataFrame(data)

# Calculate information gain for each feature
features = ['Weekend', 'Weather', 'Tired']
target = 'What to Do?'

for feature in features:
    ig = information_gain(df, feature, target)
    print(f"Information Gain for {feature}: {ig}")
```

# COnfusion matrix (calculation)

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import numpy as np

# Step 1: Create the new DataFrame
data = {
    'Weekend': ['Yes', 'No', 'No', 'No', 'Yes'],
    'Weather': ['Sunny', 'Rain', 'Rain', 'Cloudy', 'Cloudy'],  # Corrected "Claudy" to "Cloudy"
    'Tired': ['No', 'Yes', 'No', 'No', 'Yes'],
    'What to Do?': ['Go for a walk', 'TV', 'Reading', 'Go for a walk', 'Reading']
}

df = pd.DataFrame(data)

# Step 2: Encode categorical variables using Label Encoding
label_encoder = LabelEncoder()

# Encode 'Weather', 'Tired', and 'Weekend'
df['Weather'] = label_encoder.fit_transform(df['Weather'])
df['Tired'] = label_encoder.fit_transform(df['Tired'])
df['Weekend'] = label_encoder.fit_transform(df['Weekend'])

# Step 3: Define features and target variable
X = df.drop('What to Do?', axis=1)  # Features ("axis=1" is for columns)
y = df['What to Do?']  # Target variable (categorical, not encoded)

# Step 4: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# Step 5: Train the Decision Tree (considering the importance of some features)
clf = DecisionTreeClassifier(criterion='entropy')  # Use 'entropy' for ID3-like behavior
clf.fit(X_train, y_train)

# Step 6: Make predictions on the test set
y_pred = clf.predict(X_test)

# Step 7: Generate confusion matrix
cm = confusion_matrix(y_test, y_pred, labels=['Go for a walk', 'TV', 'Reading'])

# Create a DataFrame for better readability
confusion = pd.DataFrame(cm, 
                         index=['Actual Go for a walk', 'Actual TV', 'Actual Reading'], 
                         columns=['Predicted Go for a walk', 'Predicted TV', 'Predicted Reading'])

# Display the confusion matrix
print(confusion)

# Optional: Visualize the decision tree
plt.figure(figsize=(12, 8))
plot_tree(clf, filled=True, feature_names=X.columns, class_names=clf.classes_, rounded=True)
plt.title('Decision Tree using ID3 Algorithm')
plt.show()
```
