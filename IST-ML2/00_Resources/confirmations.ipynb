```python
import pandas as pd
import numpy as np
from scipy.stats import multivariate_normal

# Load the dataset from the CSV file
csv_filename = "2_homework2.csv"  # Ensure to set the correct path to your CSV file
df = pd.read_csv(csv_filename)

# Calculate means for each class
means = df.groupby('Class')[['x1', 'x2']].mean()
# Calculate covariance for each class
covariances = df.groupby('Class')[['x1', 'x2']].cov()

# Display the means and covariances
print("Means:")
print(means)
print("\nCovariances:")
print(covariances)

# Create a query vector (you can change this to any vector you want to test)
query_vector = np.array([1, 2])  # Example query
print("\nQuery Vector:")
print(query_vector)

# Calculate likelihoods for each class using multivariate normal distribution
likelihoods = {}
for cls in means.index:
    mean = means.loc[cls].values
    cov = covariances.loc[cls].values.reshape(2, 2)  # Reshape for 2D covariance
    likelihood = multivariate_normal.pdf(query_vector, mean=mean, cov=cov)
    likelihoods[cls] = likelihood

# Display results
print("\nLikelihoods:")
for cls, likelihood in likelihoods.items():
    print(f"P(X | Class={cls}) = {likelihood}")

# Determine the most probable class
most_probable_class = max(likelihoods, key=likelihoods.get)
print(f"\nMost Probable Class for the query vector {query_vector} is: {most_probable_class}")
```
